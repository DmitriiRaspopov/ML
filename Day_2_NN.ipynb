```python
train_data=mnist_train.values[:,1:]
test_data=mnist_test.values[:,1:]
```

```python
train_label=mnist_train.values[:,0]
test_label=mnist_test.values[:,0]
```

```python
print(train_data.shape, test_data.shape)
print(train_label.shape, test_label.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier=KNeighborsClassifier(n_jobs=-1)
```

```python
kn_class=knn_classifier.fit(train_data, train_label)
```

```python
test_id=1929

plt.imshow(test_data[test_id,:].reshape(28,28), cmap='Greys')
print("На рисунке представлена цифра {}".format(test_label[test_id]))
```

```python
kn_class.predict(test_data[test_id,].reshape(1,784))
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn_pred=kn_class.predict(test_data)
```

```python
accuracy_score(knn_pred, test_label)
```
