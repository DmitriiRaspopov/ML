# Метрики кластеризации

# Гомогенность и полнота

Мы хотим в конечном счете, чтобы каждому нашему кластеру соответствовал какой-то класс. И чтобы классы в кластерах не перемешивались.

Допустим у нас есть 9 классов и какое-то число кластеров.

Метрикой будет энтропия (H). Рассмотрим i-тый кластер и смотрим распределение клсса по кластерам.

В идеале энтропия должна быть для каждого кластера равна 0. Чтобы кластер был гомогенным и состоял из элеметов одного класса.

Метрики ARI and AMI делают ровным счетом то же самое и сравнивают вектор ответов кластеров с вектором классов.


# Силуэт

В отличие от других метрик, в данной метрике мы не знаем количество наших классов, и поэтому не знаем как проводить кластеризацию(по какому количеству классов).

Задача опять та же самая - чтобы объекты внутри кластеров были одинаковые(похожи) а за кластерами разные.

Чем ближе силуэт (S) к единице тем лучше(более плотная кластеризация). Интервал значений S (-1; 1).

Силуэт выдает шарообразные кластеры.

С другими формами кластеров он нам врядли сможет помочь.
